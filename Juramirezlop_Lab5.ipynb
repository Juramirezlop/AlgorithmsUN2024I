## 1. Write a code that produces the Bell numbers that count the possible partitions of a set for $n$ elements.

**Hints**:
-  Output the n-th Bell Number - [stackexchange](https://codegolf.stackexchange.com/questions/132379/output-the-n-th-bell-number).

- Bell number [wikipedia](https://en.wikipedia.org/wiki/Bell_number)






```python
def bell(n):
    bell = [0] * (n + 1)
    bell[0] = 1
    history = [1]  # Inicialmente solo tenemos el primer valor de Bell

    for i in range(1, n + 1):
        bell[i] = 0
        for j in range(i):
            bell[i] += bell[j] * coeficiente(i - 1, j)
        history.append(bell[i])

    return history

def coeficiente(n, k):
    if k == 0 or k == n:
        return 1
    else:
        return coeficiente(n - 1, k - 1) + coeficiente(n - 1, k)

# Calculating Bell numbers with history for n = 5
n = 10
bell_history = bell(n)
print("Bell numbers for n =", n, bell_history)
```

## 2.Write a code that that produces the list of all partititions of the set of $[n]=\{ 1,2,3,4 \cdots , n-1, n\}$ with $0 \le n \le 10$.

**Hints**:
- Set partitions in Python - https://stackoverflow.com/questions/19368375/set-partitions-in-python
- Generate all partition of a set - https://www.geeksforgeeks.org/generate-all-partition-of-a-set/

```python
def print_partitions(partitions):
    for partition in partitions:
        print("{" + ", ".join(map(str, partition)) + "}")

def partitions(n):
    if n == 0:
        return [[]]
    partitions_list = [[[]]] + [[] for _ in range(n)]
    for i in range(1, n + 1):
        for partition in partitions_list[i - 1]:
            partitions_list[i].append(partition + [i])
            for j in range(len(partition)):
                partitions_list[i].append(partition[:j] + [i] + partition[j:])
    return partitions_list[n]

# Generating partitions for n = 0 to 10
for n in range(11):
    print("Partitions for n =", n)
    print_partitions(partitions(n))
    print()
```

## 3. Write a code that produces the list of all partititions of the set of $[n]=\{ 1,2,3,4\cdots , n-1, n\}$ with $0 \le n \le 10$ and the sum of elements of the partition which is the a constant $\frac{n(n+1)}{2}$.

```python
from itertools import combinations

def partitions(set_):
    if not set_:
        yield []
        return
    for i in range(1, len(set_)):
        for combo in combinations(set_, i):
            remaining = set_ - set(combo)
            for partition in partitions(remaining):
                yield [list(combo)] + partition

def constant_sum_partitions(n):
    constant = n * (n + 1) // 2
    valid_partitions = []
    for partition in partitions(set(range(1, n + 1))):
        if all(sum(subset) == constant for subset in partition):
            valid_partitions.append(partition)
    return valid_partitions

# Generating partitions for n = 0 to 10
for n in range(11):
    print(f"Partitions of {set(range(1, n + 1))} with sum {(n * (n + 1)) // 2}:")
    for partition in constant_sum_partitions(n):
        print(partition)
    print()
```
